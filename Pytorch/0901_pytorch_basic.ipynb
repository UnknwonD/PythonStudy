## Reshape

```python
x = torch.randn(10, 64, 7, 7)
x.shape
```

```python
y = x.reshape(x.shape[0], -1) # (10, 64 * 7 * 7)로 변환
y.shape
```

```python
x = torch.randn(10, 28, 28, 3)
y = x.permute(0, 3, 1, 2)
x.shape, y.shape
```
